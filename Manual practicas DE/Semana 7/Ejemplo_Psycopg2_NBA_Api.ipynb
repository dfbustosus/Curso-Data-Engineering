````python
# 1. Librerias
import requests
import pandas as pd
# 2. url de api
base_url = "https://www.balldontlie.io/api/v1/players"
````

Documentación API: https://www.balldontlie.io/home.html#getting-started

````python
# 3. Hacemos el retrive de la data
response = requests.get(f"{base_url}?page={1}")
# 4. Obtener el json con la info
data= response.json()
# 5. mostrar las keys del json
print(data.keys())
# 6. Obtener el id de interes
jugadores=data.get('data')
# 7. Mostrar como viene la estructura del json
print(jugadores[0:1]) # muestra de prueba
# 8. Mostrar la cantidad de jugadores en una llamada
print(len(jugadores))
# 9. Convertir a dataframe
df=pd.DataFrame(jugadores).sort_values(by='id').reset_index(drop=True) # tabla destino
df
````

````python
# 12. Crear la tabla de destino
table_schema = f"""
    CREATE TABLE IF NOT EXISTS NBA_stats_david (
        id INT,
        first_name VARCHAR(50),
        height_feet FLOAT,
        height_inches FLOAT,
        last_name VARCHAR(50),
        position VARCHAR(50),
        weight_pounds FLOAT
    );
    """
    # Crear la tabla
cur = conn.cursor()
cur.execute(table_schema)
````

````python
from psycopg2.extras import execute_values
# 13.  Definir columnas de interes a llenar
columns = list(df.columns)
# 14. Generar los valores a insertar
values = [tuple(x) for x in df.to_numpy()]
# 15. Definir el INSERT
table_name='NBA_stats_david'
insert_sql = f"INSERT INTO {table_name} ({', '.join(columns)}) VALUES %s"
# 16. Execute the transaction to insert the data
cur.execute("BEGIN")
execute_values(cur, insert_sql, values)
cur.execute("COMMIT")
````

````python
# 17. Verificando que esta todo bien
cur.execute('SELECT * FROM NBA_stats_david')
cur.fetchall()
````

````python
# 18. Importante cerrar la conexion
cur.close()
conn.close()
````
